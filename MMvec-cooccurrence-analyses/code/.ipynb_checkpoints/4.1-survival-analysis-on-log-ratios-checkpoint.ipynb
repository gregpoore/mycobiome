```r
library(survival)
library(broom)
```

```r
days_df <- read.table('../results/tables/log_ratio_comparisons/TCGA-metadata-plus-log-ratios-by-cancer-type.tsv', sep='\t', quote="", header = TRUE)
cancer_types <- unique(days_df$TCGAStudy) 
lrs_use <- unique(days_df$log_ratio_used) 
lrs_use
```

```r
for(lr_class in lrs_use){
    
    days_df_lr_subset <- days_df[days_df$log_ratio_used == lr_class,]
    
    datalist = list()
    coxlist = list()
    i = 1
    for(cancer_type in cancer_types){
        days_df_subset <- days_df_lr_subset[days_df_lr_subset$TCGAStudy == cancer_type,]
        if(dim(days_df_subset)[1] > 5){
            if(length(unique(unlist(days_df_subset[c("log_ratio_class")]))) > 1){
        fit <- survfit(Surv(OSTime, OS) ~ log_ratio_class, data=days_df_subset)
        res <- summary(fit)
        save.df <- tidy(fit)
        save.df$TCGAStudy <- cancer_type
        datalist[[i]] <- save.df

        coxfit <- coxph(
          Surv(OSTime, OS) ~ log_ratio_class, 
            data=days_df_subset,
          ties = 'exact')
        coxfit.df <- tidy(coxfit)
        coxfit.df$TCGAStudy <- cancer_type
        coxlist[[i]] <- coxfit.df
        i = i + 1}}
    }

    big_data = do.call(rbind, datalist)
    cox_data = do.call(rbind, coxlist)
    write.csv(big_data, file = paste("../results/tables/log_ratio_comparisons/",lr_class,"-by-cancer-type-suv-os.csv", sep = ""))
    write.csv(cox_data, file = paste("../results/tables/log_ratio_comparisons/",lr_class,"-by-cancer-type-suv-cox.csv", sep = ""))

          
}
```

```r
days_df <- read.table('../results/tables/log_ratio_comparisons/TCGA-metadata-plus-log-ratios.tsv', sep='\t', quote="", header = TRUE)
cancer_types <- unique(days_df$TCGAStudy) 
lrs_use <- unique(days_df$log_ratio_used) 
lrs_use
```

```r
for(lr_class in lrs_use){
    
    days_df_lr_subset <- days_df[days_df$log_ratio_used == lr_class,]
    
    if(dim(days_df_lr_subset)[1] > 10){
        if(length(unique(unlist(days_df_lr_subset[c("log_ratio_class")]))) > 1){
    fit <- survfit(Surv(OSTime, OS) ~ log_ratio_class, data=days_df_lr_subset)
    res <- summary(fit)
    save.df <- tidy(fit)

    coxfit <- coxph(
      Surv(OSTime, OS) ~ log_ratio_class, 
        data=days_df_lr_subset,
      ties = 'exact')
    coxfit.df <- tidy(coxfit)}}

    write.csv(save.df, file = paste("../results/tables/log_ratio_comparisons/",lr_class,"-all-suv-os.csv", sep = ""))
    write.csv(coxfit.df, file = paste("../results/tables/log_ratio_comparisons/",lr_class,"-all-suv-cox.csv", sep = ""))

          
}
```
